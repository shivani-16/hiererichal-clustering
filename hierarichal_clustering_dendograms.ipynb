```python
import numpy as np
import matplotlib.pyplot as plt
#we will use make_blobs dataset to perform clustering
from sklearn.datasets import make_blobs
```

```python
#create a blob of 200 datapoints(blob-binary large object:collection of binary data)

dataset= make_blobs(n_samples=200,
                   n_features=2,
                   centers= 4,
                   cluster_std= 1.6,
                   random_state= 50)
#features means 2-dim data we are going to make
#centers means number of clusters we want
#cluster_std: standard deviation should be between 1-2
```

```python
#the data set is actually made of two parts: dataset[0] and dataset[1]
dataset
```

```python
#we want to extract only dataset[0]
points= dataset[0]
```

```python
points
```

```python
#to perform hierarichal clustering import folllowing libraries:
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering #agglomerative clustering is class which is imported from sklearn.cluster library
```

```python
#create a dendrogram as dendrogram is the key to perform agglomerative clustering
dendrogram = sch.dendrogram(sch.linkage(points,method="ward")) #sch is object and dendrogram is function, linkage is parameter
#which takes the dataset
#points is dataset and methos is: ward-ward is a method choose a pair of cluster to merge
```

```python
#to get the scatter plot of this dataset
plt.scatter(dataset[0][:,0], dataset[0][:,1])
```

```python
#perform actual clustering
#hc will perform clustering
hc= AgglomerativeClustering(n_clusters=4,affinity='euclidean',linkage='ward')
#n_clusters: will set the number of parameters
#affinity: will set to euclidean to find the distance between data points
# linkage is set to ward which minimizes the variant between clusters
```

```python
#y_hc=hc.affinity(points)
```

```python
#fit_predict predict the clusters that each data point belongs to
#now we are trying to fit the datapoints to the clustering
y_hc=hc.fit_predict(points)
```

```python
plt.scatter(points[y_hc==0,0], points[y_hc== 0,1], s=50, c='cyan', marker="*")#s is size and c is colour
plt.scatter(points[y_hc==1,0], points[y_hc== 1,1], s=50, c='blue', marker="*")
plt.scatter(points[y_hc==2,0], points[y_hc== 2,1], s=50, c='yellow', marker="*")
plt.scatter(points[y_hc==3,0], points[y_hc== 3,1], s=50, c='red', marker="*")
```
