```python
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
import numpy as np
```

```python
customer_data= pd.read_csv('hierarchical-clustering-with-python-and-scikit-learn-shopping-data.csv')
```

```python
customer_data.shape
```

```python
customer_data.head()
```

```python
data= customer_data.iloc[:,3:5].values
#[:]means keep all the rows and [3:5] means keep only 3rd and 4th column
```

```python
import scipy.cluster.hierarchy as shc
plt.figure(figsize=(10,7))
plt.title("customer dendograms")
dend= shc.dendrogram(shc.linkage(data, method='ward'))
```

```python
from sklearn.cluster import AgglomerativeClustering
cluster= AgglomerativeClustering(n_clusters=5, affinity='euclidean', linkage='ward' )
cluster.fit_predict(data)
#Since we had five clusters, we have five labels in the output i.e. 0 to 4.
```

```python
plt.figure(figsize=(10, 7))
plt.scatter(data[:,0], data[:,1], c=cluster.labels_, cmap='rainbow')
```
